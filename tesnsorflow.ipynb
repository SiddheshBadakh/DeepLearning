```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('Churn_Modelling.csv')
```

```python
X = pd.get_dummies(df.drop(['Exited', 'CustomerId'], axis=1))
y = df['Exited'].apply(lambda x: 1 if x == 'yes' else 0)
```

```python
X_train, X_test,y_train,y_test = train_test_split(X, y, test_size=0.2)
```

**1.Import Dependencies**

```python
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense  # Correct import for Dense
from sklearn.metrics import accuracy_score
```

**2.Build and Compile Model**

```python
model = Sequential()  # Initialize the model
model.add(Dense(units=32, activation='relu', input_dim=len(X_train.columns)))  # First hidden layer
model.add(Dense(units=32, activation='relu'))  # Second hidden layer
model.add(Dense(units=1, activation='sigmoid'))
```

```python
# Compile the model
model.compile(loss='binary_crossentropy', optimizer='sgd', metrics=['accuracy'])
```

**3.Fit Predict and Evalute**

```python
# Fit the model to the training data
model.fit(X_train, y_train, epochs=200, batch_size=32)
```

```python
# Make predictions
y_hat = model.predict(X_test)

# Convert probabilities to binary class labels
y_hat = [0 if val < 0.5 else 1 for val in y_hat]
```

```python
accuracy = accuracy_score(y_test, y_hat)

# Print the accuracy
print(f'Accuracy Score: {accuracy:.2f}')
```
